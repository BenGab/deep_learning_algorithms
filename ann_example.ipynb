```python
import tensorflow as tf
import numpy as np
```

```python
input_cnt = 2
output_cnt = 1
hidden_cnt = 3
```

```python
X_data = np.loadtxt("ingatlan.txt", usecols=(0, 1))
Y_data = np.loadtxt("ingatlan.txt", usecols=(2))
```

```python
X_train = X_data
Y_train = Y_data
```

```python
X = tf.placeholder(tf.float32, shape=(1, input_cnt), name="X")
Y_true = tf.placeholder(tf.float32, shape=(1, output_cnt), name="Y_true")
```

```python
W1 = tf.Variable(tf.random_uniform(shape=(input_cnt, hidden_cnt)))
B = tf.Variable(tf.random_uniform(shape=(1, hidden_cnt)))
W2 = tf.Variable(tf.random_uniform(shape=(hidden_cnt, output_cnt)))
```

```python
Y = tf.matmul(tf.sigmoid(tf.matmul(X, W1) + B), W2)
diff = (Y - Y_true) ** 2
train_op = tf.train.GradientDescentOptimizer(0.1).minimize(diff)
```

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for i in range(50):
        for t in range(X_train.shape[0]):
            _, loss_v = sess.run([train_op, diff], feed_dict={X: np.reshape(X_train[t], (1,2)), Y_true: np.reshape(Y_train[t], (1,1))})
            print("Train Loss: {}".format(loss_v))
    
    for i in range(50):
        for t in range(X_train.shape[0]):
            loss_v = sess.run([diff], feed_dict={X: np.reshape(X_train[t], (1,2)), Y_true: np.reshape(Y_train[t], (1,1))})
            print("Test Loss: {}".format(loss_v))
```
