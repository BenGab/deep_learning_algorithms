```python
import tensorflow as tf
```

Creating a Simple neural network, which uses only one neuron, and just for weights.
X is input, W is weight as a Variable, Y the output, Y_ the expected output. Loss is (Y-Y_)^2

```python
X=tf.placeholder(dtype=tf.float32, name='X')
W=tf.Variable(initial_value=tf.random_uniform([1], minval=-10.0, maxval=10.0), name='W')
Y = X*W
Y_ = tf.placeholder(dtype=tf.float32, name='Y_')
Loss = (Y-Y_) ** 2
training_op = tf.train.GradientDescentOptimizer(learning_rate=0.1).minimize(Loss)
```

Starting training session, which initializes the variables and run the training in 500 epochs

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for i in range(500):
        _, w_value, loss_value = sess.run([training_op, W, Loss], feed_dict={X: 1.0, Y_: 2.0})
        print("Loss: {}, Weight: {}".format(loss_value, w_value))
```

As it seems most ranges, does not changes the loss value. So it can cause overfitting, but the real problem is too much range or learning rate.
