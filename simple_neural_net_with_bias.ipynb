```python
import tensorflow as tf
```

As it seen earlier it will be similar work, but it extends it with bias value.

```python
X = tf.placeholder(dtype=tf.float32, name='X')
W = tf.Variable(initial_value=tf.random_uniform([1], minval=-10.0, maxval=10.0), name='W')
B = tf.Variable(initial_value=tf.random_uniform([1], minval=-10.0, maxval=10.0), name='B')
Y = X * W + B
Y_ = tf.placeholder(dtype=tf.float32, name='Y_')
Loss = (Y-Y_) ** 2
training_op = tf.train.GradientDescentOptimizer(learning_rate=0.1).minimize(Loss)
```

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for i in range(500):
        _, w_value, b_value, loss_value = sess.run([training_op, W, B, Loss], feed_dict={X: 1.0, Y_: 2.0})
        print("loss: {}, weight: {}, bias: {}".format(loss_value, w_value, b_value))
```
